## Task 1: Hangman Game

**Goal:** Create a simple text-based Hangman game where the player guesses a word one letter at a time.

### Simplified Scope:
- Use a small list of 5 predefined words.
- Limit incorrect guesses to 6.
- Basic console input/output (no graphics or audio).

### Key Concepts Used:
- `random` module
- `while` loop
- `if-else` conditional statements
- `lists` and `strings` manipulation

```python
import random

def play_hangman():
    # 1. Predefined pool of 5 words
    words_pool = ["python", "developer", "coding", "internship", "software"]
    
    # Randomly select a word from the list
    secret_word = random.choice(words_pool)
    
    # Track guessed letters
    guessed_letters = []
    
    # Limit incorrect guesses to 6
    max_attempts = 6
    incorrect_guesses = 0

    print("=== Welcome to the Hangman Game! ===")
    print(f"I have selected a secret word. You have {max_attempts} incorrect attempts allowed.")
    print("Good Luck!\n")

    # Game Loop
    while incorrect_guesses < max_attempts:
        # Display the current state of the word
        displayed_word = ""
        for letter in secret_word:
            if letter in guessed_letters:
                displayed_word += letter + " "
            else:
                displayed_word += "_ "
        
        print(f"Word to guess: {displayed_word.strip()}")
        print(f"Remaining incorrect attempts: {max_attempts - incorrect_guesses}")
        print(f"Guessed letters: {', '.join(guessed_letters) if guessed_letters else 'None'}")
        
        # Check if the player has guessed the entire word
        if "_" not in displayed_word:
            print(f"\n🎉 Congratulations! You guessed the word: '{secret_word}' correctly!")
            break

        # Get player input
        guess = input("Guess a letter: ").lower().strip()

        # Input Validation
        if len(guess) != 1 or not guess.isalpha():
            print("❌ Invalid input. Please enter a single letter.\n")
            continue
        
        if guess in guessed_letters:
            print(f"⚠️ You already guessed '{guess}'. Try another letter.\n")
            continue

        # Add the guess to our tracker list
        guessed_letters.append(guess)

        # Check if the guess is in the secret word
        if guess in secret_word:
            print(f"✅ Good job! '{guess}' is in the word.\n")
        else:
            incorrect_guesses += 1
            print(f"❌ Oops! '{guess}' is not in the word.\n")

    # If the player runs out of attempts
    if incorrect_guesses >= max_attempts:
        print("Game Over! 😢")
        print(f"The secret word was: '{secret_word}'")

# Run the game
play_hangman()
```
