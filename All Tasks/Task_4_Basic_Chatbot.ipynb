# Task 2: Basic Chatbot
### Goal: Build a simple rule-based chatbot
### Key Concepts Used: if-elif, functions, loops, input/output

This Jupyter Notebook contains a fully functional, rule-based chatbot designed to meet the criteria of **Task 2**. It handles basic user input, processes it efficiently by ignoring capitalization and extra spaces, and runs continuously in a loop until the user decides to exit.

```python
def get_chatbot_response(user_input):
    """
    Processes user inputs and returns a structured string response.
    Checks for predefined keywords and matches them with appropriate replies.
    """
    # Capitalization aur extra spaces ka masla hal karne ke liye lowercase aur strip kiya
    clean_input = user_input.strip().lower()
    
    # --- Rule-Based logic using if-elif-else ---
    if clean_input in ["hello", "hi", "hey", "aoa", "assalam o alaikum"]:
        return "Hi there! How can I help you today? 😊"
        
    elif clean_input in ["how are you", "how are you?", "kaise ho", "how's it going"]:
        return "I'm doing great, thank you for asking! How about you? 🚀"
        
    elif clean_input in ["fine", "good", "i am fine", "i'm fine", "sab set hai"]:
        return "That's wonderful to hear! Keep smiling! ✨"
        
    elif clean_input in ["what is your name?", "what is your name", "naam kya hai"]:
        return "I am ChatBuddy, your virtual assistant. 🤖"
        
    elif clean_input in ["bye", "goodbye", "exit", "allah hafiz"]:
        return "Goodbye! Have a fantastic day ahead! 👋"
        
    else:
        return "I'm sorry, I didn't quite catch that. Could you please try again? 🤔"
```

```python
def start_chatbot():
    """
    While loop use karke chatbot ka interactive session run karta hai.
    """
    print("=" * 50)
    print("          WELCOME TO CHATBUDDY (v1.0)          ")
    print("   (Type 'bye' or 'exit' to end the conversation)  ")
    print("=" * 50)
    
    # --- Loop for continuous conversation ---
    while True:
        # Taking input from user
        user_msg = input("\nYou: ")
        
        # Getting response
        bot_reply = get_chatbot_response(user_msg)
        
        # Displaying the response
        print(f"Chatbot: {bot_reply}")
        
        # Stop condition
        if user_msg.strip().lower() in ["bye", "goodbye", "exit", "allah hafiz"]:
            break

# Program start karne ke liye
if __name__ == "__main__":
    start_chatbot()
```
