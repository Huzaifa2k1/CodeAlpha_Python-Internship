### Goal:
Build a stock portfolio tracker that calculates total investment based on manually defined stock prices using a pre-defined dictionary, handles user inputs, and optionally saves the summary to a file.

---

### Step 1: Define Stock Prices (Dictionary)
We use a predefined dictionary `STOCK_PRICES` to store the ticker symbols and their respective prices. We also define a helper function to display these options beautifully.

```python
# Hardcoded stock prices (Dictionary)
STOCK_PRICES = {
    "AAPL": 180.0,
    "TSLA": 250.0,
    "MSFT": 420.0,
    "GOOGL": 170.0,
    "AMZN": 185.0
}

def display_available_stocks():
    """Prints the list of available stocks and their hardcoded prices."""
    print("\n=================================")
    print("    AVAILABLE STOCKS & PRICES    ")
    print("=================================")
    for stock, price in STOCK_PRICES.items():
        print(f"  • {stock:<6} : ${price:>6.2f}")
    print("=================================\n")
```

### Step 2: Main Portfolio Tracker Logic
This block runs the interactive loop. It prompts the user to input stock names and quantities, performs inputs checks, calculates totals, and provides an option to export the output to a text file (`portfolio_receipt.txt`).

```python
def main():
    portfolio = {}
    display_available_stocks()
    
    # 1. User Input Loop
    while True:
        stock_input = input("Enter stock symbol to buy (or type 'calculate' to finish): ").upper().strip()
        
        if stock_input == 'CALCULATE':
            break
            
        # Check if stock exists in our dictionary
        if stock_input not in STOCK_PRICES:
            print(f"❌ Error: '{stock_input}' is not in the system. Please select from the list.\n")
            continue
            
        # Input validation for quantity
        try:
            quantity = int(input(f"How many shares of {stock_input} do you want to buy? "))
            if quantity <= 0:
                print("❌ Quantity must be greater than 0.\n")
                continue
            
            # Add or update quantity in the portfolio dictionary
            portfolio[stock_input] = portfolio.get(stock_input, 0) + quantity
            print(f"✅ Added {quantity} shares of {stock_input} to your portfolio.\n")
            
        except ValueError:
            print("❌ Invalid input! Please enter a whole number for quantity.\n")

    # If user didn't purchase anything
    if not portfolio:
        print("\nPortfolio is empty. Exiting tracker.")
        return
 
    # 2. Calculation & Summary Generation
    print("\n" + "="*45)
    print("               PORTFOLIO SUMMARY               ")
    print("="*45)
    
    total_investment = 0.0
    summary_lines = []
    
    for stock, qty in portfolio.items():
        price = STOCK_PRICES[stock]
        cost = qty * price
        total_investment += cost
        
        line = f" {stock:<6} | {qty:>3} shares x ${price:>6.2f} = ${cost:>8.2f}"
        print(line)
        summary_lines.append(line)
        
    print("-"*45)
    total_line = f" Total Portfolio Value: ${total_investment:>21.2f}"
    print(total_line)
    print("="*45)
    
    # 3. File Handling (Saving the result to a text file)
    save_choice = input("\nDo you want to save this summary to a text file? (yes/no): ").lower().strip()
    if save_choice in ['yes', 'y']:
        file_name = "portfolio_receipt.txt"
        try:
            with open(file_name, "w") as file:
                file.write("=============================================\n")
                file.write("           STOCK PORTFOLIO SUMMARY           \n")
                file.write("=============================================\n")
                for line in summary_lines:
                    file.write(line + "\n")
                file.write("---------------------------------------------\n")
                file.write(total_line + "\n")
                file.write("=============================================\n")
            print(f"💾 Success! Saved to '{file_name}'.")
        except IOError:
            print("❌ Error: Could not save file.")

# Run the portfolio tracker
main()
```
